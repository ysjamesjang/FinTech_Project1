```python
lumber_prices_2015= lumber_prices[lumber_prices.index >= '2015-01-01']
lumber_prices_2015
```

```python
lumber_prices_2015_mean = lumber_prices_2015.reset_index()
lumber_prices_2015_mean["Value"] = lumber_prices_2015_mean["Value"].astype(int)
lumber_prices_2015_mean.dtypes
```

```python
mean_2015_2019 = lumber_prices_2015_mean.loc[lumber_prices_2015_mean["Date"]<="2019-12-01"]
mean_2015_2019
mean_2015 = mean_2015_2019['Value'].mean()
mean_2015
```

```python
mean_2020_present = lumber_prices_2015_mean.loc[lumber_prices_2015_mean["Date"]>"2019-12-01"]
mean_2020 = round(mean_2020_present['Value'].mean(),2)
mean_2020
```

```python
def plot_prices(lumber_price, ylabel, title):
    lumber_price = pd.DataFrame(lumber_prices_2015_mean["Value"])
    Lumber_complete_plot = lumber_price.hvplot.line(
    label = title,
    ylabel = ylabel,
    rot = 90,).opts(width=900, height=350)
    
    hline = hv.HLine(212.19)
    hline.opts(
    color='red',
    line_dash='dashed',
    line_width=2.0,)
    
    hline1 = hv.HLine(285.69)
    hline.opts(
    color='black',
    line_dash='dashed',
    line_width=2.0,)
    return Lumber_complete_plot*hline*hline1
```

```python
plot_prices("lumber_prices_2015_mean", "lumber_price", "Mean prices pre and during pandemic")
```

```python
import pandas as pd

midwest_df = pd.read_csv("Resources/midwest.csv")
avg_col = midwest_df.loc[:, 'Illinois (IL)':'South Dakota (SD)']
midwest_df['Midwest'] = avg_col.mean(axis=1)
midwest_df = midwest_df[['Date','Midwest']]

northeast_df = pd.read_csv("Resources/northeast.csv")
avg_col = northeast_df.loc[:, 'Connecticut (CT)':'Pennsylvania (PA)']
northeast_df['Northeast'] = avg_col.mean(axis=1)
northeast_df = northeast_df[['Date','Northeast']]

south_df = pd.read_csv("Resources/south.csv")
avg_col = south_df.loc[:, 'Delaware (DE)':'Texas (TX)']
south_df['South'] = avg_col.mean(axis=1)
south_df = south_df[['Date','South']]

west_df = pd.read_csv("Resources/west.csv")
avg_col = west_df.loc[:, 'Arizona (AZ)':'Washington (WA)']
west_df['West'] = avg_col.mean(axis=1)
west_df = west_df[['Date','West']]

combined_df = midwest_df.merge(northeast_df, on='Date').merge(south_df, on='Date').merge(west_df, on='Date')
combined_df.head()
```
